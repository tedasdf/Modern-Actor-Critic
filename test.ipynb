```python
import torch

def get_flat_params(model):
    return torch.cat([p.data.view(-1) for p in model.parameters()])
```

```python
from utilities.neuralnet import Actor


act_net = Actor(
    3,10, 4
)


flat_param = get_flat_params(act_net)

print(flat_param.shape)
```
